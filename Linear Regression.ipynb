```python

#Linear Regression Strach
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as st
st.use("ggplot")
data = pd.read_csv("C://Users/NayakD/Desktop/headbrain.csv")
data.head()

X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
i = 0
j = 0
for k in range(n):
    i += (X[k]-mean_x)*(Y[k]-mean_y)
    j += (X[k]-mean_x) ** 2

m=i/j
c = mean_y - (m*mean_x)
print(m,c)
# plot
max_x=np.max(X)
min_x = np.min(X)

x = np.linspace(min_x-100,max_x+100,1000)
y = c+m*x
plt.plot(x,y,color="#000000", label="line")
plt.scatter(X,Y,label="scatter")
plt.legend()
plt.show()
# r square value = 1 - ((x-xpred)**2/(x-xmean)**2)
sr =0
s = 0
for i in range(n):
    ypred = m*X[i] + c
    #print(ypred)
    s += (Y[i]-mean_y) ** 2
    sr += (Y[i]-ypred) ** 2
r2 = 1-(sr/s)
print(r2)
```

# Using Sklearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#Rank1 convertion
X = X.reshape((n,1))
#model_creation
reg = LinearRegression()
#fitting data to model
reg = reg.fit(X,Y)
#Y_pred
ypred = reg.predict(X)
# R2 score
r2 = reg.score(X,Y)
print(r2)
```
